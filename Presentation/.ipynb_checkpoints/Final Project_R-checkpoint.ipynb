#### Introduction and Problem Definition

            Tanzania started their free water project for all in 1971 after they gained independence. As a socialist country, it had water extraction and distribution firmly under government control through Village Water Committee. By 1992, policy makers realized that their current water policy was not working as 50% of the rural population still had no access to water. The government started privatizing water extraction and distribution stations. Its goal was to convert the Village Water Committee to a regulatory body rather than an active managing partner of water distribution channels. 
           With this big picture as a backdrop, this paper's problem statement comes from DrivenData. DrivenData is an online host of data science problems for NGOs in areas such as microfinance default modeling, blood donation prediction to text analysis of budget line items for education districts. The problem set in this paper is on predicting faulty water pumps in Tanzania. More accurate prediction could lead to better allocation of resources in fixing these water pumps and continued water supply for many Tanzanians relying on them. The data is provided by Ministry of Water in Tanzania and Taarifa, a open-sourced platform for crowd reporting of infrastructure related problems.

#### Modeling Overview

           To achieve a model that generalizes well for our competition, we carry out the following steps.
               1. Pick Model
                  There are different types of models. If the response variable is continuous, we use regression     models.  If the response variable is categroical, we use classification models. Another differential is between parametric and non-parametric models. Parametric models have limited number of independent variables that act as predictors. It is mainly used for predictive problems that are non-complex and have clear prediction cases. Examples of parametric models are linear regression and logistic regression Non-parametric models, on the other are for more complex predictive problem sets. They do not a limited number of independent variables, but, can be considered to be infinite. These models are to pick naunces in the data set and works well with non-clear cut prediction problems. Examples of non-parametric models are KNN and decision trees. 
               2. Split Data
                    Wth training on known data set comes the problem of overfitting. Overfitting is when the model picks up both the noise and the signal in the data and works well only for that limited data set. We use the following two methods to resolve, overfitting issues. 
                   1. Train/Test split - Here, we basically split the data to training and testing data set. In sckit-learn the ratio is 70-30. 
                   2. Cross Validation - Cross Validation is the Train/Test split on steroids. One of the disadvantage of Train/Test split is that there is a high variance with whatever evaluation metrics we apply to the Train/Test split model. The data is split randomly and in certain instances: the Training data set A could lead to higher accuracy than Training data set B. To rein in the variance, we essentially split the data set into however many K - folds we want and apply train-test split by having each fold as testing data set once with the others as training data. Higher numbers of folds are computationally expensive so, there needs to be a balance between gaining accuracy and computation cost. If we need a quick sandbox model, Train/Test split is sufficient. Another usage of Cross-Validation is in model and feature selection. Cross Validation function usually comes with the evaluation metrics parameter that enables us to easily compare the results we desire between different model choice as well as feature choice.
               3. Train Model
                   This part is where we set up the data to be put into the model of our choice. We will choose the features and dummy the data if necessary to be fitted into the model.
               4. Test Model
                   This part is for evaluating the performance of our model on out of sample data/testing data. There are several metrics we can use but for our purposes, the following three are all we need.
                   1. Classification Accuracy - this metrics is pretty straightforward. This measures how many classes have been predicted accurately. Higher score usually corresponds to a better model.
                   2. Confusion Matrix - this is a more customized metrics. Here, we are dealing with sensitivity and specificity scores. Sensitivity refers to the rate of prediciting a true case when it is actually true. Specificity refers to the rate of prediciting a false case when it is actually false. For instance, in spam and not spam, assume no spam is 0 and spam is 1. If we are more concerned by accurately identifying spams and catching them then, we would increase sensitivity value over specificity. Increasing one decreases the other. Ultimately, choosing one over the other depends on the priority of the business case.
                   3. ROC/AUC - Receiver Operating Characteristics and Area Under the Curve for ROC allows us to pick the best threshold for our sensitivity and 1-specificity, or indirectly, specificity trade-offs. ROC curve plots each corresponding sensitivity and 1-specificity value for each probability threshold for binary classification. AUC is the area of the curve with 1.0 being the best model and 0.5 being a model that indicates no distinction between sensitivity and specificity.
               5. Parameter Tuning - Each of the model that parameters that can be adjusted. We are adding in some codes to go through those parameter changes to get the best results for each model.
               6. Choose best model - We basically compare all the models we have trained and tested our data and pick the best one.
               7. Train on all data - We will test this on the complete data set instead of split one as we will be using unseen data subsequently.
               8. Make prediction on unseen data from competition - These are truly out of sample data that we will be using.

#### Logistic Regression - An initial look

           For a first-pass model, logistic regression is used as it is simple to implement and train without scaling features. It also provides well-calibrated probabilities which helps with AUC metrics score. The competition focuses on classification accuracy score but, the focus of this paper would be on AUC.From an operational standpoint, it is more useful to know which pumps are not working or needs fixing than which pumps are working. Therefore, the Functional Needs Repair and Not Functional classes can be lumped together as one. The goal is to label them as True Positives and increasing the sensitivity of the prediction which indirectly, is increasing AUC. 

         Below, we have converted the multinomial response variables into a binary class, dummied the predictors and prepared for independent X variables and response y to be supplied into the logistic model.

```python
waterjoined['NeedsRepair_Other'] = waterjoined.status_group.map({'functional': 0, 'non functional' :1, 'functional needs repair': 1 })
```

```python
management_dummies = pd.get_dummies(waterjoined.management, prefix = 'Management')
region_dummies= pd.get_dummies(waterjoined.region, prefix = 'Region')
waterpoint_dummies = pd.get_dummies(waterjoined.waterpoint_type, prefix = 'WaterPoint')
```

```python
WaterX = pd.concat([management_dummies, region_dummies, waterpoint_dummies], axis = 1)
```

```python
X = WaterX
y = waterjoined['NeedsRepair_Other']
```

          Train/Test split is applied and logistic model is imported for fitting. Subsequently, the test data set is applied to the model.

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e9)
logreg.fit(X_train,y_train)
```

```python
y_pred_class = logreg.predict(X_test)
```

        To access our model with classification accuracy, we must first establish a baseline. The baseline or null accuracy can be calculated by assuming that the model only predicts the majority class precisely. In our case, our model's accuracy is at 66.31% compared to 54.24% for the null which is good for a first-pass.

###Regularization

            Regularization is a technique used on linear models to decrease error due to variance. It is used when the model is complex or have a sizable independent variable. This technique decreases the size of the coefficients and in lasso (l1) regularization's case, it removes some features altogether. As a result, lasso does automatic 'feature selection'. In the initial modeling, we have picked features for the model after some data exploration and have not thrown in more variables without some testing as it may cause overfitting. With lasso, a higher number of potentially 'relevant' variables can be put into the model as lasso would take care of those features that are 'irrelevant'.
            
             As mentioned in the data cleaning section above, there are duplicate categorical variables that need to be examined and removed. The Appendix section contains a detailed exploration of duplications and the reasons behind its removal. Out of 39 independent features, 12 are removed due to repetition and granularity. From the remaining 25,the 3 selected in the initial model along with 10 new features are added to the model. The features that are excluded pertain to different markers of the location of the water point such as longitude, latitude, num_private, basin, subvillage, region, region_code, distrinct_code, lga, ward whose information are more or less conveyed in the region feature already. The features included each singal different types of information about the waterpoint such as whether there is a public meeting or not, the type of extraction used, the aount of water available, the height of where the water pump is, the installer of the pump etc. Some of these variables may not correlate at all to whether a water pump functions or not. This is not our concern since lasso will remove it if it is so. Some features that are left behind are date_recorded and construction year. Those were not included because date_recorded conveys nothing much about the pump. The construction year, on the other hand has variability and the suscipion is that the age of pump would be more predictive which is explored later.

####Tuning Regularization

            We are going to use two new functions to get the optimal penalty and regularization method. First, pipeline allows us to sequennce multiple estimators (transformers and classifiers). In our case, we have standard scaler as a transformer and logistic regression as a classifer to be applied sequentially. The next step is setting up GridSearchCV. We have to establish the range of the penalty C which we have from ten power negative 3 to 2 across both lasso (l1) and ridge (l2). The k-fold is set for 4 and the scoring function is roc_auc so that we can maximize sensitivity score.

```python
#pipeline of Standard Scaler and Logistic Regression
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(StandardScaler(),LogisticRegression())
```

###Decision Trees

            After pushing the limits of a parametric model, the next model I have decided to use is decision tree. It is a non parametric model that is highly interpretable as it resembles human decision making process. The goal of decision tree is to use a greedy method in figuring out the nodes that would give us the highest class purity for a given tree depth using the gini index. Assume a binary classification problem. The gini index for each node tells us how much it has differentiated the testing sample into its two response classes correctly.  A gini index of 0.5 would indicate that the variable has not really divided the sample into its response classes. The resulting two fork contains a mixture of both classes in each. A gini index of 0 on the other hand incidates that the two forks from the variable contain only each class. At each level of the tree for each node, the algorithm would search for the next variable with the least gini index until a maximum tree depth has been reached.
            The setting of code for decision tree is pretty straightforward as laid out before. It follows the standard procedure of spliting the data and then, fitting into the model with a specified tree depth.

**Results**

            The accuracy results are better than the initial non regularized model but it is not better than the regularized model with the standing result at 71% compared to 77% from the regularized model. The AUC score is significantly less at 70% compared to 84% from the regularized model. Another feature of the decision tree is its ability to tell us the most important variable for classification. In this particular case, a Waterquantity being dry or not tells us most whether that water point is function or not. It makes common sense as a waterpoint supplied by very little water would not be working. Next, we shall see if decision tree can perform better after tuning. 

```python
X_train, X_test, y_train,y_test = train_test_split(X,y, random_state=1)
```

####Tuning Decision Trees

            Tuning decision tree is a matter of finding the maximum tree level depth at which the evaluation metrics score that we use is at its maxium. The other caveat is to use k-fold cross validation in order that we do not overfit on training data. 
            Below we running the decision tree from tree depth range of 1 to 8. At each tree depth, we are using k fold of 5 and getting an average gini index score out of it. Finally, we are plotting that average gini score by the tree depth.

**Improved Results**

            The AUC score that we are using as metric performs better as the tree depth go deeper. However, it seems to be tapering of around tree depth of 7. It is still not at the same level as regularization but, this is worth investigation later on for non-parametric models. 

```python
from sklearn.cross_validation import cross_val_score
max_depth_range = range(1,8)
Gini_Scores = []
for depth in max_depth_range:
    treereg = DecisionTreeClassifier(max_depth = depth, random_state =1)
    Gini_Score = cross_val_score(treereg,X,y, cv=5,scoring='roc_auc')
    Gini_Scores.append(np.mean(Gini_Score))
```

```python
plt.plot(max_depth_range, Gini_Scores)
plt.xlabel('max_depth')
plt.ylabel('roc_auc (higher is better)')
```

Feature engineering is the act of creating new variables out of existing one that may be more predictive. We will apply some data exploration to those new variable before putting it into a model. The question becomes which variable to pick. It is a matter of applying domain knowledge and critical thinking to the existing set of variables.

**Age of Pump Cleaning and Exploration**

            The component of the first new feature, construction year is made of 32% zero values. The zero values are as good as NA and have to be imputed. The optimal way is to neutralize that data set by randomly distributing the years with respect to the non-zero year distribution. Such a task requires more custom code without knowing the possibility of the effectiveness of this new feature. Therefore, a given imputation method 'most frequent' is used to replace the zero values. The other possible methods were mean and median. Mean was not used since it would be odd to have non-integer years. Median and  most frequent would not be that different as it would expand upon one particular year or the other.
            After the imputation, we find 17 records to have negative years. It is not very suprising since some date recorded values can be lower in value (older from present time) than construction year. 17 is a negligible number so, we can move ahead.
